Gráfico de setores de um conjunto qualitativo de dados:

```python
import matplotlib.pyplot as plt
import numpy as np

Y = ["Bom", "Ruim", "Ótimo", "Regular", "Regular", "Ótimo", "Ótimo","Bom", "Ótimo", "Bom", "Ótimo"]

values = np.sort(np.unique(Y))
freq = np.zeros(len(values))
explode = (0.1, 0, 0, 0)
ind = 0

for i in values:

  counter = 0

  for j in range(0, len(Y)):

    if Y[j] == i:

      counter += 1

  freq[ind] = counter
  ind += 1

fig1, ax1 = plt.subplots(figsize = (10, 6))
ax1.pie(freq, explode = explode, labels = values, autopct = '%1.1f%%', shadow = True, startangle = 90)
ax1.axis('equal')
plt.show()
```
