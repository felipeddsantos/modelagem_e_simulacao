Média de um conjunto quantitativo de dados:

```
import numpy as np 

X = [1, 1, 1, 2, 2, 3, 4, 5, 5, 5]
mx = np.mean(X)

print('X:', X)
print('Média de X:', mx)
```

Mediana do mesmo conjunto de dados:

```
md = np.median(X)

print('X:', X)
print('Mediana de X:', md)
```

Média e mediana do mesmo conjunto de dados com valor extremo:

```
Y = [1, 1, 1, 2, 2, 3, 4, 5, 5, 100]

print('Y:', Y)
print('Nova média:', np.mean(Y))
print('Nova mediana:', np.median(Y))
```

Dessa forma, podemos concluir que a média é muito sensível a valores extremos, enquanto que a mediana é mais robusta
