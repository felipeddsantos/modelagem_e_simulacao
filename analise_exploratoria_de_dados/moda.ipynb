Moda de um conjunto quantitativo de dados:

```
from statistics import mode

X = [1, 2, 3, 1, 2, 3, 4, 1, 3, 6, 4, 1, 4, 4, 5, 0, 9, 9, 8, 7, 1, 2, 1, 1]
m = mode(X)

print('Moda de X:', m)
```

Moda de um conjunto qualitativo de dados:

```
Y = ["red", "blue", "blue", "red", "green", "red", "red"]

print('Moda de Y:', mode(Y))
```
