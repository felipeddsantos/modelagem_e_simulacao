Cálculo da média, variância, IQR e amplitude de dois conjuntos de dados:

```
from scipy.stats import iqr
import numpy as np

def variancia(X):
    
    m = np.mean(X)
    N = len(X)
    s = 0
    
    for i in np.arange(0, len(X)):
    
        s += (X[i] - m) ** 2
    
    s /= (N - 1)
    
    return s

X = [0, 0, 1, 1, 18]
Y = [4, 4, 4, 4, 4]

mx = np.mean(X)
my = np.mean(Y)

print('X:', X)
print('Média de X:', mx)
print('Variância de X:', variancia(X))
print('IQR de X:', iqr(X))
print('Amplitude de X:', np.max(X) - np.min(X))

print('\n')
print('Y:', Y)
print('Média de Y:', my)
print('Variância de Y:', variancia(Y))
print('IQR de Y:', iqr(Y))
print('Amplitude de Y:', np.max(Y) - np.min(Y))
```
