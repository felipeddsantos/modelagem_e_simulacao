A distribuição geométrica é uma distribuição de probabilidade discreta que descreve a probabilidade de ocorrer um sucesso em $k$ tentativas, sendo $p$ a probabilidade de ocorrer sucesso. Uma variável aleatória $X$ segue a distribuição geométrica se sua função densidade de probabilidade for dada por:

image.png

A seguir, será realizada uma simulação da distribuição geométrica.

```python
#Simulação da distribuição geométrica
#Para esse exemplo, a probabilidade de sucesso será de 0.35
#Serão realizados 100000 experimentos (extração de variáveis aleatórias)

#Declaração das bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt

#Probabilidade de sucesso
p = 0.35

#Número de pontos extraídos da distribuição 
n = 100000

#Obtenção do vetor de variáveis aleatórias geradas pela distribuição geométrica
X = np.random.geometric(p, size = n)
k = np.arange(0, np.max(X))

#Comparação das esperanças (teórica e experimental)
print("Valor esperado (teoria):", 1 / p)
print("Valor esperado (simulação):", np.sum(X) / len(X))

#Plotagem do gráfico da distribuição
count, bins, ignored = plt.hist(X, bins = k, density = True, color = '#0504aa', alpha = 0.7, rwidth = 0.9)
plt.xlabel('k', fontsize = 15)
plt.ylabel('P(k)',fontsize = 15)
plt.show(True)
```

A seguir, será executado um conjunto de $n = 1000$ simulações, onde cada simulação consistirá em $k$ lançamentos de uma moeda até que ocorra um sucesso (cara), com probabilidade $p = 0.4$ de ocorrência.

```python
#Simulação de lançamento de uma moeda com k - 1 fracassos consecutivos (coroa) e um sucesso (cara)
#Nesse exemplo, serão realizadas 1000 simulações, com probabilidade 0.4 de sucesso (cara)

#Declaração das bibliotecas necessárias
from random import seed
from matplotlib import pyplot as plt
import numpy as np

#Chance de sucesso (cara)
p = 0.4

#Armazenamento do número de lançamentos
X = [] 

#Número de simulações
ns = 1000

#Laço para executar as ns simulações
for i in range(0, ns):

    #Número de lançamentos
    k = 0

    #Indica um fracasso
    s = False

    #Laço que executa o lançamento da moeda enquanto sair fracasso (coroa)
    while s == False:

        #Geração de número aleatório entre 0 e 1
        r = np.random.uniform()
        
        #Se sucesso, incrementa o número de lançamentos e finaliza o laço
        if(r <= p):

            s = True
            k = k + 1
        
        #Se falha, incrementa o número de lançamentos e continua o laço
        else:

            k = k + 1

    #Armazena a quantidade de lançamentos necessários para o sucesso na simulação corrente
    X.append(k)

k = np.arange(0, np.max(X))

#Comparação das esperanças (teórica e experimental)
print("Valor esperado (teoria):", 1 / p)
print("Valor esperado (simulação):", np.sum(X) / len(X))

#Comparação das variâncias (teórica e experimental)
print("Variância (teoria):", (1 - p) / p ** 2)

#Função para o cálculo da variância experimental
def Variance(X):
    
    n = len(X)
    E2 = 0
    E = 0
    
    for i in range(0, len(X)):
    
        E2 = E2 + X[i] ** 2
        E = E + X[i]

    E2 = E2 / n
    E = E / n
    V = E2 - (E) ** 2
    
    return V
    
print("Variância (simulação):", Variance(X))

#Plotagem do gráfico da distribuição
count, bins, ignored = plt.hist(X, bins = k, density = True, color = '#0504aa', alpha = 0.7, rwidth = 0.9)
plt.xlabel('k', fontsize = 15)
plt.ylabel('P(k)',fontsize = 15)
plt.show(True)
```

Como foi observado, em ambas as simulações, a esperança e a variância experimental possuem um valor próximo àqueles obtidos teoricamente. Esses valores seriam exatos se fossem realizados um número infinito de experimentos.
