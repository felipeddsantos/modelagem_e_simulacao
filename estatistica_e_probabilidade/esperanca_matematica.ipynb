Nas simulações a seguir, será calculada a esperança matemática de uma série de problemas de maneira experimental e teórica, comparando os valores obtidos. 

Problema I: lançamento de um dado honesto.

Questão: Qual é a esperança do valor obtido no lançamento do dado?

Se um dado honesto for lançado e $X$ designar o valor obtido, a esperança teórica é dada por:

$E(X) = 1/6 + 2/6 + 3/6 + 4/6 + 5/6 + 6/6 = 3.5$. 

A esperança experimental será calculada a seguir.

```python
#Experimento I: simulação do problema I para o cálculo da esperança
#Nesse caso, o dado será lançado 10000 vezes

#Declaração das bibliotecas necessárias
from random import seed
from random import randint

#Semente do gerador de números aleatórios
seed(1)

#Média
mu = 0

#Quantidade de lançamentos
N = 10000

for n in range(0, N):

    #Valor do dado
    mu += randint(1, 6)

print('Valor médio de', N, 'lançamentos:', mu / N )
```

Problema II: Um fabricante produz peças tais que $10 \%$ delas são defeituosas e $90 \%$ são perfeitas. Se uma peça defeituosa for produzida, o fabricante perde $1,00$ real, enquanto que uma peça não-defeituosa lhe dá um lucro de $5,00$ reais. 

Questão: Qual a esperança do valor do lucro obtido pelo fabricante?

Se $X$ for o lucro líquido por peça, a esperança teórica é dada por: 

$E(X)=5*0.9-0.1 = 4,40$

A esperança experimental será calculada a seguir.

```python
#Experimento II: simulação do problema II para o cálculo da esperança
#Nesse caso, serão realizadas 10000 simulações

#Declaração das bibliotecas necessárias
from random import seed
import numpy as np

#Semente do gerador de números aleatórios
seed(1) 

#Número de simulações
N = 10000

#Lucro por peça
X = 0 # lucro por peca

#Probabilidade de produzir uma peça perfeita
p = 0.9

#Lucro para peça perfeita
lucro = 5

#Prejuízo pra peça imperfeita
prejuizo = 1

for n in range(0, N):

    #Número aleatório de 0 a 1
    r = np.random.uniform()

    #Se o número gerado é menor do que p, peça perfeita
    if(r <= p):

        X += lucro

    else:

        X -= prejuizo

print('Valor médio do lucro:', X / N, 'reais por peça')
```

Problema III: Um caça-níquel tem dois discos que funcionam independentemente um do outro. Cada disco tem 10 figuras: 4 maçãs, 3 bananas, 2 peras e 1 laranja.Uma pessoa paga R$ 80,00 e aciona a máquina. Se aparecer:

* 2 maçãs, ganha R\$ 40,00
* 2 bananas, ganha R\$ 80,00
* 2 peras, ganha R\$ 140,00
* 2 laranjas, ganha R\$ 180,00

Se aparecer frutas diferentes, ele não ganha nada. 

Questão: Qual a esperança do valor do lucro obtido pelo jogador?

Se $X$ for o lucro líquido por jogo, a esperança teórica é dada por: 

$E(X)= 40*0.4^2 + 80*0.3^2 + 140*0.2^2 + 180*0.1^2 - 80 = -59,00$ 

OBS: Não joguem esse jogo!

A esperança experimental será calculada a seguir.


```python
#Experimento III: simulação do problema III para o cálculo da esperança
#Nesse caso, serão realizadas 10000 simulações

#Declaração das bibliotecas necessárias
import numpy as np
from random import seed
import random

#Semente do gerador de números aleatórios
seed(100)

frutas = ["Maca", "Maca", "Maca", "Maca", "Banana", "Banana", "Banana", "Pera", "Pera", "Laranja"]
pagamento = 80

#Números de jogadas
N = 10000

#Lucro por jogada
X = 0

for n in range(0, N):

    disco1 = random.choice(frutas)
    disco2 = random.choice(frutas)
  
    if(disco1 == disco2):
  
        if(disco1 == "Maca"):
  
            X += 40
  
        else:

            if(disco1 == "Banana"):
          
                X += 80
          
            else:
          
                if(disco1 == "Pera"):
          
                    X += 140
          
                else:
          
                    if(disco1 == "Laranja"):
          
                       X += 180
    
    X -= pagamento

print('Valor médio do lucro em uma única jogada:', X / N)
```

Após a realização dos experimentos, notamos que os valores das esperanças obtidos nas simulações são bem próximos dos valores obtidos analiticamente. Para um valor exato, seriam necessários um número de simulações que tende ao infinito.
