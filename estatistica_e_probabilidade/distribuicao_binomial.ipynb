A distribuição binomial é a distribuição de probabilidade discreta que calcula a probabilidade de ter exatamente $k$ sucessos em $n$ tentativas, com $p$ a probabilidade de sucesso. A distribuição binomial possui uma função densidade de probabilidade dada por:

image.png

A seguir, será realizada uma simulação da distribuição binomial.

```
#Simulação da distribuição binomial (lançamento de uma moeda)
#Para esse exemplo, a probabilidade de sucesso (cara) será de 0.4 com 500 lançamentos de moeda 
#Serão realizados 10000 simulações (extração de variáveis aleatórias)

#Declaração das bibliotecas necessárias
from random import seed
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import binom
import math

#Semente do gerador de números aleatórios
seed(100)

#Quantidade de lançamentos
n = 500

#Probabilidade de sucesso (cara)
p = 0.4

Pk = np.zeros(n)
vk = np.arange(0, n)

#Quantidade de simulações
ns = 10000

#Realiza as ns simulações
for j in range(0,ns):

    #Número de sucessos (caras)
    S = 0
    
    #Realiza n lançamentos
    for i in range(0, n):

        #Gera um número aleatório entre 0 e 1
        r = np.random.uniform() 

        #Se sucesso, incrementa a quantidade de sucessos
        if(r <= p):

            S = S + 1

    Pk[S] = Pk[S] + 1

#Normalização da distribuição de probabilidade
Pk = Pk / sum(Pk)

#Plotagem do gráfico da distribuição
plt.figure(figsize = (10, 6))
plt.xlim(0.8 * np.min(vk[Pk > 0]), 1.2 * np.max(vk[Pk > 0]))
plt.bar(vk, Pk, label = 'Simulada')

#Cálculo da curva teórica
Pkt = np.zeros(n + 1)
vkt = np.arange(0, n + 1)
for k in range(0, n + 1):

    Pkt[k] = (math.factorial(n) / (math.factorial(n - k) * math.factorial(k))) * (p ** k) * (1 - p) ** (n - k)

plt.plot(vkt, Pkt, 'r--', label = 'Teórica')
plt.xlabel('k', fontsize = 15)
plt.ylabel('P(k)',fontsize = 15)
plt.legend()
plt.show(True)
```

As curvas teórica e experimental convergiriam no caso de um número infinito de simulações.
