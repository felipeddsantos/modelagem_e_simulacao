Suponha que você esteja participando de um jogo com três portas e que atrás de uma delas tenha um carro como prêmio. O apresentador do jogo pede a você que escolha uma porta. Se você escolher a porta que tem o carro, você ganha o carro.

Entretanto, após você escolher sua porta, o apresentador abre uma outra porta do jogo e mostra que aquela porta contém um bode. O apresentador, então, pergunta: “Você quer mudar de porta?”. Há apenas duas portas sobrando, aquela que você escolheu primeiro e aquela que não foi aberta.

A questão que se coloca é: Qual a probabilidade de você ganhar o prêmio, se mudar de porta? É mais vantajoso mudar de porta ou permanecer com a mesma?

A seguir, será realizada uma simulação do problema de Monty-Hall

```python
#Simulação do problema de Monty-Hall
#Nesse caso, serão realizadas 5000 simulações

#Importação das bibliotecas necessárias
import numpy as np
from matplotlib import pyplot as plt
import random
from random import seed

#Semente do gerador de números aleatórios
seed(100)

#Quantidade de portas do jogo
portas = [1,2,3]

#Quantidade de simulações
Nmax = 5000

#Armzenamento da quantidade de vitórias
P = []

#Armazenamento do número de jogos
vN = []

for N in np.arange(1, Nmax, 10):

    vitoria = 0
    derrota = 0
  
    for i in range(0, N):

        #Coloca o carro em uma porta
        C = random.choice(portas)

        #Seleciona uma porta
        X = random.choice(portas)

        #Se o carro não está na porta selecionada, o apresendador irá abrir outra porta que não tem o carro, a porta será trocada e ganharemos
        if(C != X):

            vitoria = vitoria + 1

    P.append(vitoria / N)
    vN.append(N)

#Plotagem do gráfico da simulação e do cálculo teórico
plt.figure(figsize = (12, 8))
plt.plot(vN, P, 'ro-', label = 'Simulação: Muda a porta')
plt.axhline(y = 2 / 3, color = 'red', label = 'Probabilidade teórica com mudança de porta')
plt.ylim(0, 1.05)
plt.legend()
plt.show(True)
```

No caso do problema de Monty-Hall, o cálculo teórico informa que a probabilidade de sucesso caso mudemos de porta é igual a 2/3 (aproximadamente 66.67%). Logo, concluimos que de fato é vantajoso mudar de porta. A simulação realizada acima nos mostra que, com o aumento do número de experimentos (jogos realizados), a quantidade de vitórias no caso da troca de porta converge para 2/3, onde, no limite, há uma convergência total. Isto é, alguém que participa do jogo infinitas vezes ganhará o carro em 2/3 das partidas.
