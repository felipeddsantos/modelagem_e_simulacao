A distribuição hipergeométrica é uma distribuição de probabilidade discreta que descreve a probabilidade de $k$ sucessos em $r$ retiradas, sem reposição, de uma população de tamanho $N$ que contém exatamente $N1$ sucessos, sendo cada retirada um sucesso ou um fracasso. Uma variável aleatória $X$ segue a distribuição hipergeométrica se sua função densidade de probabilidade for dada por:

image.png


A seguir, será realizada uma simulação da distribuição hipergeométrica.

```python
#Simulação da distribuição hipergeométrica
#Para esse exemplo, haverão 20 elementos do tipo sucesso, 30 elementos do tipo fracasso e 25 elementos selecionados
#Serão realizados 100000 experimentos (extração de variáveis aleatórias)

#Declaração das bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt

#Quantidade de elementos do tipo I (sucessos)
N1 = 20

#Quantidade de elementos do tipo II (fracassos)
N2 = 30

#Número de objetos selecionados
r = 25

#Número de pontos extraídos da distribuição 
n = 100000

#Obtenção do vetor de variáveis aleatórias geradas pela distribuição hipergeométrica
X = np.random.hypergeometric(N2, N1, r, n)
k = np.arange(0, np.max(X))

#Comparação das esperanças (teórica e experimental)
print("Valor esperado (teoria):", N2 * N1 / (N1 + N2))
print("Valor esperado (simulação):", np.sum(X) / len(X))

#Plotagem do gráfico da distribuição
count, bins, ignored = plt.hist(X, bins = k, density = True, color = '#0504aa', alpha = 0.7, rwidth = 0.9)
plt.xlabel('k', fontsize = 15)
plt.ylabel('P(k)',fontsize = 15)
plt.show(True)
```

Como foi observado, a esperança experimental possui um valor próximo àquele obtido teoricamente. Esses valores seriam exatos se fossem realizados um número infinito de experimentos.
